## Standardization

In the following dataframes, we had to standardize litteral objects in order to optimize merging. We therefore created a function (standardize column) that removes accent, coma and plurals from the strings, and also puts them in lower case. We also removed special strings like what was following '√† l'exclusion de' (meaning 'except') because it created a lot of wrong merges. 

```python
#We want to remove anything that is excluded from the CPF6 code
def remove_exclusion(string): 
    if ("a l'exclusion" in string):
        return string[:string.index("a l'exclusion")-1]
    else: return(string)
```

```python
#Standardise everything to be singular. Some excpetions need to be handled
def remove_plural(Produit):
    if(('pois' in Produit)|('mais' in Produit)|('radis' in Produit)|('cassis' in Produit)|('panais' in Produit)|('brebis' in Produit)|('frais' in Produit)
      |('bruxelles' in Produit)):    
        return Produit
    else: 
        Produit = Produit.replace("s ", " ")
        return Produit[:-1] if Produit.endswith("s") else Produit
```

```python
# remove accent, lower case, coma and plurals 
def convert(input_str):
    output_str = unicodedata.normalize('NFKD', str(input_str.encode('utf-8')))
    output_str = output_str.encode('ASCII', 'ignore').lower().replace(",","")
    return remove_exclusion(remove_plural(output_str)).replace(" nda", "")

def convert_unicode(unicodestring):
    return remove_plural(unidecode.unidecode(unicodestring).lower().replace(",",""))

def standardize_column(df, colu):
    df[colu] = df[colu].apply(convert_unicode)
    return df
```

## Add months in which the products are in season to cpf6 dataframe

 First, we want to associate the products categories (cpf6_s, 2.e) with the months in which they are in season in France. To do that, we created a function that associates the cpf6 dataframe with the seasons in seasonal products (seasonnal_products_s, 2.g). 

```python
def add_seasons_to_cpf6(row):
    tab = cpf6_s[cpf6_s['Name'] == row['Products']] # If matching names --> directly adds months to cpf6_s dataframe
   
    if(len(tab)==0):
        tab = cpf6_s[cpf6_s['Name'].str.contains(row['Products'])] # If there's no exact match --> see if the seasonnal products 
                                                                   # name is contained in the cpf6_s products names, and merge
                                                                   # if yes. 
    
    if(len(tab)!=0):
    # Merging with other dataFrame
        for j in range(len(tab)):
            if(cpf6_s['Month'].iloc[tab.index.values[j-1]]==0):
                cpf6_s['Month'].iloc[tab.index.values[j-1]] = row['Month']
                
            # Adding new month only if not already in the list
            elif((row['Month'] in cpf6_s['Month'].iloc[tab.index.values[j-1]]) == False): 
                cpf6_s['Month'].iloc[tab.index.values[j-1]] = cpf6_s['Month'].iloc[tab.index.values[j-1]] \
                + ',' + row['Month']
    
```

```python
# This is a function that split the values of a column and expand so the new DataFrame has one split
# value per row. Filters rows where the column is missing.
def tidy_split(df, column, sep=',', keep=False):
    indexes = list()
    new_values = list()
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        values = presplit.split(sep)
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    return new_df
```

## Add cpf6 code to the production dataframe

We next wanted to associate the production dataframe (productions_s, 2.f) products with the CPF6 code from cpf6_s (2.e). 

```python
#iteration on production_s
def add_cpf6_to_production(row):
    
    tab = cpf6_s[cpf6_s['Name']==row['Name']] # If matching names --> directly adds cpf6 to productions_s
   
    if(len(tab)==0):
        tab = cpf6_s[cpf6_s['Name'].str.contains(row['Name'])] # If there's no exact match --> see if the productions_s 
                                                               # name is contained in the cpf6_s products names, and merge
                                                                # if yes. 
    
    if(len(tab)!=0):
    #Merging with other dataFrame
    
        str_cpf6 = 0
        for j in range(len(tab)):

            if(str_cpf6 ==0):
                str_cpf6 = cpf6_s['CPF6'].iloc[tab.index.values[j-1]]
            else: 
                if (type(str_cpf6)!= str):
                    string = str_cpf6.astype('str') 
                else: string  = str_cpf6 # converting code to str
                str_cpf6 =  string + ',' + cpf6_s['CPF6'].iloc[tab.index.values[j-1]]
        return str_cpf6
    else: return 0
```

## Production per month

Importations are registered for each month in the importations dataframe. It's registered for the whole year in the production dataset. We assumed that the production of a product in France could only occur during the months in which the product is 'in season', and we also assumed that the production is the same for all these months. We therefore divided total production among the months in which the product is in season. 

```python
#Selecting columns of interest
Month_production = Seasonnal_production[['CPF6', 'Value','month_idx']]
Month_production.head()
```

```python
def length(list_): 
    return(float(len(list_)))

def convert_list_to_string(list_): 
    list_ = [str(i) for i in list_]
    str_ = ','.join(list_)
    return str_
```

```python
# Converting months list to string, in order to split. Also saving season length to divide production by the number of months
# in which the product is in season. 

Month_production['season_len']= Month_production['month_idx'].apply(length)
Month_production['month_idx']= Month_production['month_idx'].apply(convert_list_to_string)
Month_production = Month_production.dropna()
Month_production["Value"] = Month_production.Value.astype("float")
Month_production.head()
```

```python
# If multiple products with the same cpf6
Month_production = Month_production.groupby(['CPF6','month_idx', 'season_len']).sum().reset_index()

#Dividing by the number of month of the season to have the production/month
Month_production['Value'] = Month_production['Value']/Month_production['season_len'] 

#Split the dataframe by month to merge it with importation data
Month_production = tidy_split(Month_production,'month_idx' , sep=',').drop('season_len', axis = 1)
Month_production["month_idx"] = Month_production["month_idx"].astype(int)

Month_production.head()
```

```python
#Sample n_iter times the dataframe 'df' with replacement and calculate each time the median value of its column
#returns an array containing the median value of each bootstrap sample
def bootstrap_mean(row):
    df = pd.Series(row)
    
    n_iter = 100
    mean_stats=[None]*n_iter
    
    for i in range(n_iter):
        sample_i = df.sample(n=len(df),replace=True) #create a bootstrap sample
        mean_i = sample_i.mean()
        mean_stats[i]= mean_i
    return mean_stats
```

> *Is the importated mass of a product for a given month significantly higher than imported mass during the rest of the year?*

To compare mensual import to annual import for each month and for every product, we created a function that performs a T-test and that says if the mensual import is significantly higher than the importation over the year. 

```python
def is_significantly_higher_than_mean(row):
    alpha = 0.05
    Ttest_indResult= sp.stats.ttest_1samp(row['All_year_mass'],row['Month_mass'])
    real_pvalue= Ttest_indResult.pvalue/2
    
    if((row['Month_mass']>np.mean(row['All_year_mass']))&(real_pvalue < alpha)):
        return True
    else: 
        return False
```

## Distribution of mean distance per month

For each month, we want to display boxplots of the distance traveled by products, computed by bootstrap. 
This would allow us to discriminate month where the distance traveled by products are significantly lower or higher than the mean distance over the year.

```python
#compute the bootstraped distances per month
distance_distribution = importations.groupby('Month')\
                                    .agg(lambda x: tuple(x)).applymap(list)\
                                    .reset_index().drop(['CPF6', 'Year', 'Mass', 'Country', 'cca2'], axis = 1)
distance_distribution['bootstrapped_distance'] = distance_distribution['dist_to_fr'].apply(bootstrap_mean)
distance_distribution.head()
```

The bootstraping allow us to say if the distances for a month are significantly different than the mean distance over the year. For a particular month, we can compute a **95% confidence interval** with the boxplot: if the mean distance over the year is not contained in this interval, the distances for this month are significantly different than the other months.

```python
means = distance_distribution.bootstrapped_distance.apply(np.mean)
mean_of_mean = means.mean()

def is_significantly_different(row):
    if ( (np.percentile(row, 2.5)>mean_of_mean) | (np.percentile(row, 97.5)<mean_of_mean) ):
        return 'rgb(242, 219, 46)'
    else :
        return 'rgb(200,200,200)'
    
distance_distribution["significant"] = distance_distribution.bootstrapped_distance.apply(is_significantly_different)
distance_distribution
```

```python
#Boxplots with plotly
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data = [{
    'name':months[i],
    'y': distance_distribution.bootstrapped_distance.values[i], 
    'type':'box',
    'marker':{'color': distance_distribution.significant[i]}
    } for i in range(12)]
layout = dict(title = 'Boxplot of bootstrapped mean distances of importations per month')
fig = fig = dict(data=data, layout=layout)
py.iplot(fig, filename="boxplot_per_month")
```

**Correlation between distance and mass importations**

```python
#Spearman correlation coefficient computed by bootstraping

#Sample n_iter times the dataframe 'df' with replacement and calculate each time the Spearman coefficient between two columns.
#returns an array containing the Spearman coefficient of each bootstrap sample
def bootstrap_spearman(df, n_iter):
    spearman_coeffs=[None]*n_iter
    
    for i in range(n_iter):
        sample_i = df.sample(n=len(df)+50,replace=True) #create a bootstrap sample
        spearman_i = spearmanr(sample_i.Mass, sample_i.dist_to_fr)
        spearman_coeffs[i]= spearman_i
    return spearman_coeffs

def boxplot(coeffs, p_value):
    #Spearman boxplot
    median = np.median(coeffs)
    median_label = str(median.round(2))
    ax = sns.boxplot(x=coeffs, orient='v', width=0.2, notch=True)
    ax.text(0,median+0.004, median_label,
           horizontalalignment='center', color='w')
    plt.ylabel('Spearman correlation coefficient')
    
    #Printing confidence intervals for coefficients and p-values
    print('95% confidence interval for Spearman Coefficient: [', np.percentile(coeffs, 2.5).round(4)\
        , ',', np.percentile(coeffs, 97.5).round(4), ']')
    print('95% confidence interval for Spearman p-value: [', np.format_float_scientific(np.percentile(p_value, 2.5),precision= 2)\
        , ',', np.format_float_scientific(np.percentile(p_value, 97.5),precision= 2), ']')
```

```python
spearman_bootstrapped = bootstrap_spearman(mass_import, 1000)
coeffs_spearman = [i[0] for i in spearman_bootstrapped]
p_values_spearman = [i[1] for i in spearman_bootstrapped]
#Spearman boxplot
boxplot(coeffs_spearman, p_values_spearman)
plt.suptitle('Boxplot of bootstrapped distribution for Spearman coefficients')
```

According to this computation of Spearman correlation coefficient, the mass of the imported products is significantly negatively correlated with the traveled distance.

## 5.7/ Correlation how far from season/distance

We want to know to what extent a product is imported from further when it is not its season in France. 
To do that, we compute the time difference between the importation month and the nearest month where this product would be in season in France.  
    For example, if salad are imported in February whereas their season is in summer and begins in June, it would be imported 4months before its right season. We have to take in count that the order of months is circular year after year, so if a product is imported in December wheras its season begins in January, the time difference will be 1 month (and not 11).

```python
def distance_from_season(row):
    month = np.append(np.full(len(row['Season_Month']), row['Importation_Month']),np.full(len(row['Season_Month']),12+ row['Importation_Month']))
    Seasons = np.append(row['Season_Month'], row['Season_Month'])
    return np.min(abs(month - Seasons))
```

```python
#compute the time difference for each product in the dataframe 'impotration_season'
importations_season['dist_from_season'] = importations_season.apply(distance_from_season, axis = 1)
importations_season = importations_season.sort_values('dist_to_fr')
importations_season.head()
```

Products coming from distant countries where the climate is different from France are likely to be imported while it is not their season in France. For example, the countries in the south hemisphere could provide products only available in summer in France. Therefore, we expect the time difference 'dist_from_season' to be somehow linked with the distance from France of origin countries.

We evaluate this by looking at the Spearman correlation between the distance traveled by the product and the distance from the season. 

```python
sp.stats.spearmanr(importations_season['dist_to_fr'], importations_season['dist_from_season'])
```

The distance of importation is not correlated with the distance from the season. We import products from the same distance, independantly from how far we are from the season. 

```python
sns.jointplot(x="dist_to_fr", y="dist_from_season", data=importations_season);
```

We can see that no matter how far we are from the season, we have the same distances: This means that we constantly import from the same countries.
